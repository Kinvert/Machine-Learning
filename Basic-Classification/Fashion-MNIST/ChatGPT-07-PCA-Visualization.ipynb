```python
# Import the necessary libraries
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

# Load the Fashion-MNIST dataset and prepare the data for PCA
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# Create a PCA model and fit it to the training data
pca = PCA(n_components=2)
pca.fit(x_train)

# Transform the training and test data into the principal components
x_train_pca = pca.transform(x_train)
x_test_pca = pca.transform(x_test)

# create a scatter plot of the training data in the principal components space
plt.figure(figsize=(10, 8))
plt.scatter(x_train_pca[:, 0], x_train_pca[:, 1], c=y_train)

# add labels to the plot
for i, txt in enumerate(y_train):
    plt.annotate(txt, (x_train_pca[i, 0], x_train_pca[i, 1]),
                 fontsize=12,
                 fontweight='bold',
                 alpha=0.0)

# set the plot to interactive mode
plt.interactive(True)

# show the plot
plt.show()
```
